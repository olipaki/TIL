```python
import requests

# 단일 리소스 조회
url = "https://jsonplaceholder.typicode.com/posts/3"
response = requests.get(url)
print("단일 조회 결과:", response.json())

# 전체 목록 조회
list_url = "https://jsonplaceholder.typicode.com/posts"
list_response = requests.get(list_url)
print("전체 게시글 개수:", len(list_response.json()))


#실습 추가 내용
list_data = list_response.json()
for post in list_data[:5]: # 딱 5개만
    print(f"글 번호 {post['id']}: {post['title']}")
#실습 추가 내용
list_data = list_response.json()
for post in list_data[:3]: 
    print(f"글 번호 {post['id']}: {post['userId']}")

# 뒤에서부터 5개만 가져오고 싶을 때 (가장 최근 글 5개)
for post in list_data[-5:]: 
    print(f"글 번호 {post['id']}: {post['title']}")
```
