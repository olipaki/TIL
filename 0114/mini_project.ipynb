```python
## 게임의 캐릭터 운영 규칙

### 1. 캐릭터는 이름과 역할을 가지고 태어납니다.
우리 게임의 모든 캐릭터는 이름표를 달고, 자기가 무엇을 하는 사람인지 역할을 정해서 태어납니다. 
이름이 없거나 역할이 정해지지 않은 상태로는 게임 세상에 발을 들일 수 없습니다.

### 2. 어떻게 싸우는지는 역할에 따라 결정됩니다.
모든 캐릭터에게 똑같이 공격하라고 명령을 내려도, 그 사람이 맡은 역할에 따라 행동이 달라집니다.

- 전사 역할을 맡은 사람에게 명령하면 본능적으로 검을 휘두르며 나갑니다.
- 마법사 역할을 맡은 사람에게 명령하면 자연스럽게 마법을 주문처럼 외우며 나갑니다.
캐릭터가 손에 무엇을 들었느냐와 상관없이, 그 사람이 평생 배워온 싸움의 방식이 나오는 것입니다.

### 3. 도구(무기)는 아무나 들 수 없습니다.
이 세상에는 검이나 지팡이 같은 다양한 도구들이 있습니다.
각 도구에는 그 도구만이 가진 필살기가 하나씩 숨겨져 있죠.
하지만 아무나 이 도구를 집어 들 수 있는 건 아닙니다.

- 각 역할마다 이런 종류의 도구만 써도 좋다는 허가 리스트가 있습니다.
- 만약 전사에게 지팡이를 쥐여주려고 하면, 
전사는 나는 이런 건 배운 적이 없어서 못 써요라며 거절합니다. 
오직 허락된 도구인 검을 줄 때만 기쁘게 받아들여 손에 쥡니다.

### 4. 무기를 들어야만 필살기를 씁니다.

캐릭터에게 무기에 숨겨진 기술을 써보라고 시켰을 때의 상황입니다.

- 손에 무기가 있다면: 그 무기에 적혀 있는 화려한 필살기를 사용합니다.
예를 들어 검이라면 강력한 베기 같은 기술이 나갑니다.
- 맨손이라면: 지금 내 손에 든 게 없는데 무엇을 쓰나요라고 당황해합니다.
즉, 무기를 제대로 장착했을 때만 가능한 행동입니다.

---

### 실제 상황 예시 (아더의 이야기)

1. 탄생: 아더라는 이름의 전사가 태어났습니다.
2. 도구 장착 시도 1: 아더에게 마법사가 쓰는 지팡이를 줬더니, 아더는 이건 내 역할에 맞지 않아요라며 거절합니다. 아더의 손은 여전히 비어있습니다.
3. 도구 장착 시도 2: 이번엔 검을 줬습니다. 아더는 이건 제가 잘 다루는 거죠라며 꽉 쥡니다.
4. 기본 공격: 아더에게 공격해라고 하자, 전사답게 검을 휘두르며 달려나갑니다.
5. 필살기 사용: 아더에게 무기 기술 써봐라고 하자, 손에 든 검의 필살기인 연속 베기를 선보입니다.

# 설계도

## OOP 기반 게임 캐릭터 시스템 설계 및 구현

### 1. 요구 사항

**[캐릭터]**

- 모든 캐릭터는 **`이름`, `직업`, `무기**` 속성을 가집니다.
- 캐릭터 생성 시 반드시 **이름**과 **직업**이 지정되어야 합니다.
- 캐릭터는 `공격하기()`와 `무기스킬사용()` 기능을 가집니다.

**[직업]**

- 직업은 **추상 클래스**로 설계하며, 모든 직업은 다음을 포함합니다.
- `공격 방식`: 직업에 따라 출력되는 메시지가 달라야 합니다.
- `장착 가능 무기 리스트`: 해당 직업이 사용할 수 있는 무기 타입들을 저장합니다.
- **구현 클래스:** '전사', '마법사'를 생성하세요.

**[무기]**

- 무기는 **추상 클래스**로 설계하며, 각 무기는 고유한 `스킬 이름`을 가집니다.
- **구현 클래스:** '검', '지팡이'를 생성하세요.

---

### 2. 핵심 로직

### 다형성을 활용한 공격 시스템

- 캐릭터의 `공격하기()`를 호출하면, 현재 장착된 **직업의 공격 방식**이 실행되어야 합니다.
- (예: 전사라면 "검을 휘두릅니다", 마법사라면 "마법을 시전합니다")

### 합성(Composition)을 활용한 장착 시스템

- 캐릭터는 `무기장착(무기객체)` 메서드를 가집니다.
- **제한 사항:** 직업에서 허용하지 않는 무기를 장착하려 할 경우, "이 직업은 해당 무기를 장착할 수 없습니다"라는 에러 메시지를 출력하고 장착되지 않아야 합니다.
- 올바른 무기를 장착했을 때만 캐릭터의 무기 슬롯에 저장됩니다.

### 무기 스킬 실행

- 캐릭터의 `무기스킬사용()`을 호출하면, 현재 장착된 **무기의 고유 스킬**이 실행되어야 합니다.
- 만약 무기를 장착하고 있지 않다면 "장착된 무기가 없습니다"라는 메시지를 출력합니다.

---

### 3. 테스트 시나리오 (Test Case)

구현이 완료되면 다음 시나리오가 정상 동작하는지 확인하세요.

1. **전사** 캐릭터 '아더' 생성.
2. '아더'에게 **지팡이** 장착 시도 (실패해야 함).
3. '아더'에게 **검** 장착 (성공해야 함).
4. '아더'의 `공격하기()` 호출 (전사의 공격 방식 출력).
5. '아더'의 `무기스킬사용()` 호출 (검의 스킬 출력).
6. **마법사** 캐릭터 '멀린'을 생성하여 동일한 과정 반복 테스트.

---
```

```python
# 공격하기 : 휘두릅니다
# 무기스킬사용 : 검의 스킬을 사용합니다.
# 아더 - 전사 - 검 
# 멀린 - 마법사 - 지팡이

# 캐릭터(몸체): 이름표를 달고 있고, 내부에 '직업' 칸과 '무기' 칸이 비어있어요.
# 직업(부품 1): "나는 전사야", "나는 마법사야"라는 정보를 캐릭터에게 알려줘요.
# 무기(부품 2): 캐릭터의 손에 들려있는 실제 물건이에요.

# 2단계: "허가 리스트 확인하기" (핵심 로직)
# 이 문제에서 가장 중요한 건 **"전사가 지팡이를 들 수 있는가?"**를 확인하는 거예요.
# 이걸 코드로 만들 때는 이렇게 생각하세요.
# 무기에는 **'타입'**이 있어요. (예: "검", "지팡이")
# 직업에는 **'허가 리스트'**가 있어요. (예: 전사=["검"], 마법사=["지팡이"])
# 확인 절차: 무기를 들려고 할 때, 무기의 타입이 직업의 허가 리스트 안에 들어있는지 확인(if문)하면 돼요!

# --- 1. 무기 설계도 ---
class attact:
    def __init__(self,name,skill_name) :
        self.name = name 
        self.skill_name = skill_name # 무기 이름 (검)

# --- 2. 직업 설계도 ---
class Job:
    def __init__(self,job_name,allowed_attact):
        self.job_name=job_name
        self.allowed_attact = allowed_attact # 허가 리스트 ["검"]

    def attact_style(self):
        pass # "공격 방식은 나중에 정하자!" (자리를 지키는 용도)
    # class Warrior(Job):
    # def attack_style(self):
    #     return "검을 휘두릅니다!"  # 여기서 pass를 지우고 진짜 내용을 써요. 이게 '오버라이딩'이에요!
class Warrior(Job):
    def __init__(self):
        super().__init__("전사",["검"]) #전사는 검만 사용
    def attack_style(self):
        return "검을 휘두릅니다" #전사만 공격
class Mage(Job):
    def __init__(self):
        super().__init__("마법사",["지팡이"]) #마법사는 지팡이만 사용
    def attack_style(self):
        return "마법 주문을 외칩니다"#마법사만 주문
    
# --- 3. 캐릭터 설계도 (메인 몸체) ---
class Character:
    def __init__(self, name, Job):
        self.name = name
        self.Job = Job 
        self.attact = None #처음에는 없음

    def okay_attact(self,attact):
        #무기 체크
        if attact.name in self.Job.allowed_attact:
            self.attact = attact
            print(f"{self.name}:{attact.name}을 장착")
        else:
            print(f"{self.name}:제 역할({self.Job.job_name})에 맞지 않아요")

    def do_attact(self):
        print(f"{self.name}이 {self.Job.attact_style()}")
    def use_skill(self):
        if self.attact:
            print(f"{self.name}이 {self.attact.name}의 필살기 [{self.attact.skill_name}] 사용")
        else:
            print(f"{self.name}:무기가 없습니다.")

 # --- 테스트 실행 ---
arthur = Character("아더", Warrior())
soma = Character("소마", Mage())
sword = attact("검", "연속 찌르기")
staff = attact("지팡이", "휘루루")

arthur.okay_attact(staff) # 실패해야 함
arthur.okay_attact(sword) # 성공해야 함
arthur.do_attact()        # 전사 공격 방식 출력
arthur.use_skill()        # 검의 스킬 출력

soma.okay_attact(staff) # 
soma.okay_attact(sword) # 
soma.do_attact()        # 전사 공격
soma.use_skill()        # 스킬
```
