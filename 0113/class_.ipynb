```python
### 문제 2
전자제품을 나타내는 `Electronics` 클래스와 이를 상속받는 `Smartphone`, `Fan` 클래스를 작성하세요.
1. `Electronics` 클래스는 `brand`를 속성으로 가지며, `power_on(self)` 메서드에서 "[brand] 제품의 전원이 켜집니다."를 출력합니다.
2. `Smartphone` 클래스는 `super()`를 사용하여 부모의 `brand`를 초기화하고, 추가로 모델명(`model`)을 속성으로 가집니다.
3. `Smartphone` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여, 부모의 기능을 먼저 실행한 후 "모델 [model]이 부팅을 시작합니다."를 추가로 출력하세요.
4. `Fan` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여, 부모의 기능을 먼저 실행한 후 "바람이 붑니다"를 추가로 출력하세요.
```

```python
class Electronics:
    def __init__(self, brand):
        self.brand = brand # 삼성, LG 같은 브랜드 이름을 저장해요.
# `Electronics` 클래스는 `brand`를 속성으로 가지며, 
# `power_on(self)` 메서드에서 "[brand] 제품의 전원이 켜집니다."를 출력합니다.
    def power_on(self):
        print(f"{self.brand} 제품의 전원이 켜집니다.")

# `Smartphone` 클래스는 `super()`를 사용하여 부모의 `brand`를 초기화하고, 
# 추가로 모델명(`model`)을 속성으로 가집니다.

class smartphone(Electronics):
    def __init__(self,brand, model):
        super().__init__(brand) # super()는 "부모님(Electronics)한테 가서 brand 저장 좀 해
        self.model = model  # 스마트폰만 가진 '모델명'을 따로 저장해요.
##############################################################################

# `Smartphone` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여,
# 부모의 기능을 먼저 실행한 후 "모델 [model]이 부팅을 시작합니다."를 추가로 출력하세요.

    def power_on(self):
        super().power_on()# 부모님이 만든 "전원이 켜집니다"를 먼저 실행!
        print(f"모델{self.model}이 부팅을 시작합니다.")# 그 다음 내 할 일 하기!

# `Fan` 클래스에서 `power_on(self)` 메서드를 오버라이딩하여, 
# 부모의 기능을 먼저 실행한 후 "바람이 붑니다"를 추가로 출력하세요.
class Fan(Electronics):
    def power_on(self):
        super().power_on()# 부모님의 "전원이 켜집니다"를 먼저 실행!
        print(f"바람이 붑니다.") # 선풍기

##############################################################
# 스마트폰 만들기
phone = smartphone("애플","아이폰 15")
phone.power_on()
# 출력시
# 애플 제품의 전원이 켜집니다. (부모님 기능)
# 모델 아이폰 15이 부팅을 시작합니다. (내 기능)

phone = smartphone("애플","아이폰 15")
phone.power_on()

# 선풍기 만들기
my_fan = Fan("신일")
my_fan.power_on()
# 출력시
# 신일 제품의 전원이 켜집니다. (부모님 기능)
# 바람이 붑니다. (내 기능)
```
