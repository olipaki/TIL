```python
a = 3

def func():
    print(a)

func()
```

```python
b = 4
def func2():
    b = 5 # 밖의 b = 4를 변경한게 아니라, 함수 공간에 b를 새로 만든 것
    print(b)
func2()
print(b)
```

함수 내부에서 외부의 값에 접근이 가능(사용 가능)
변경은 불가능

```python
lst = [1, 2, 3]

def func3():
    lst[0] = 100
    print(lst)

func3()
print(lst)
```

```python
lst = [1, 2, 3]

def func3():
    lst = [100, 2, 3]
    print(lst)

func3()
print(lst)
```
