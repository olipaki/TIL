### 문제 1
나이를 입력받아 다음 조건에 해당하면 "무료 입장입니다"를 출력하고, 그렇지 않으면 "유료 입장입니다"를 출력하세요.
- 조건: 나이가 7세 이하이거나 65세 이상인 경우

```python
# 나이를 입력 받아야 한다.
age = int(input()) # str으로 입력이 받아집니다.
# if 나이가 7세 이하이거나 65세 이상인 경우:
if (age <= 7) or (age >= 65): 
    print('무료 입장')
else:
    print('유료 입장')
```

### 문제 4
아이디(`user_id`)와 비밀번호(`user_pw`)를 입력받습니다.
- 아이디가 `"admin"`이고 비밀번호가 `"1234"`인 경우: "로그인 성공" 출력
- 아이디가 `"admin"`이지만 비밀번호가 틀린 경우: "비밀번호가 틀렸습니다" 출력
- 아이디가 틀린 경우: "존재하지 않는 아이디입니다" 출력

```python
user_id = input()
user_pw = input()

if user_id == 'admin' and user_pw == '1234':
    print('로그인 성공')
if user_id == 'admin' and user_pw != '1234':
    print('비밀번호 틀림')
if user_id != 'admin':
    print('존재하지 않는 id')

# id 확인
if user_id == 'admin':

    # pw 확인
    if user_pw == '1234':
        print('로그인 성공')
    else:
        print('비밀번호 틀림')
        
else:
    print('존재하지 않는 id')
```
