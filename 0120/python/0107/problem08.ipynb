### 문제 1
입력받은 점수에 따라 합격/불합격을 판별하고 메시지를 반환하는 함수 `get_grade_result`를 작성하세요.

**조건:**
1. 점수가 60점 이상이면 "합격"을 반환합니다.
2. 점수가 60점 미만이면 "불합격"을 반환합니다.
3. 만약 점수가 0점보다 작거나 100점보다 크면 "잘못된 점수"를 반환하세요.

```python
score = 30
if score < 0 or score > 100 :
    print('잘못된 점수')
else:
    if score >= 60:
        print('합격')
    else:
        print('불합격')

    
```

```python
def get_grade_result(score):
    if score < 0 or score > 100 :
        return '잘못된 점수'
    else:
        if score >= 60:
            return '합격'
        else:
            return '불합격'

    
```

```python
def get_grade_result(score):
    
    # 비정상적인 data에 대한 처리를 우선 한다.
    if score < 0 or score > 100 :
        return '잘못된 점수'
    
    # 조건에 대한 결과를 준다.
    if score >= 60:
        return '합격'
    
    # 기본값이다.
    return '불합격'

    
```

### 문제 2
입력받은 숫자가 홀수인지 짝수인지 판별하여 문자열을 반환하는 함수 `check_even_odd`를 작성하세요.

**조건:**
1. 숫자가 짝수이면 "짝수"를 반환합니다.
2. 숫자가 홀수이면 "홀수"를 반환합니다.

```python
def check_even_odd(n):
    if n % 2 == 0:
        return '짝수'
    return '홀수'
```

### 문제 3
두 개의 숫자와 연산자(+, -)를 입력받아 계산 결과를 반환하는 아주 간단한 계산기 함수 `simple_calc`를 작성하세요.

**조건:**
1. `op`가 "+"이면 두 수의 합을 반환합니다.
2. `op`가 "-"이면 두 수의 차를 반환합니다.
3. 그 외의 연산자가 들어오면 "지원하지 않는 연산"이라는 메시지를 반환하세요.

```python
num1 = 10
num2 = 20
op = '+'
def simple_calc(num1, num2, op):
    # available_op = {'+', '-'}

    # if op not in available_op:
    #     return "지원하지 않는 연산"

    if op == '+':
        return num1 + num2
    elif op == '-':
        return num1 - num2
    
    return "지원하지 않는 연산"
```

```python

# 테스트
print(simple_calc(10, 5, "+")) # 결과: 15
print(simple_calc(10, 5, "-")) # 결과: 5
print(simple_calc(10, 5, "*")) # 결과: "지원하지 않는 연산"
```

### 문제 4
이름 리스트를 입력받아 이름이 3자 이상인 사람만 골라 새로운 리스트로 반환하는 함수 `filter_long_names`를 작성하세요.

**조건:**
1. `for`문을 사용하여 리스트의 각 이름을 확인합니다.
2. `len()` 함수를 사용하여 이름의 길이를 확인합니다.

```python
name_list = ["김철수", "이영희", "박씨", "최도날드"]

def filter_long_names(name_list):
    # 주석을 적어주면 좋아요.
    
    new_list = []
    for name in name_list:
        if len(name) >= 3:
            new_list.append(name)
    return new_list

def filter_long_names(name_list):
    # 함수 내부의 구현은 강사가 말하는 '못생긴 코드'여도 괜찮습니다.
    return [name for name in name_list if len(name) >= 3]
```

```python

# 테스트
name_list = ["김철수", "이영희", "박씨", "최도날드"]
print(filter_long_names(name_list)) # 결과: ["김철수", "이영희", "최도날드"]
# print([name for name in name_list if len(name) >= 3])
```

### 문제 6
장바구니에 담긴 상품들의 최종 결제 금액을 계산하는 함수 `calculate_checkout_price`를 작성하세요.

**조건:**
1. `membership_grade`가 'VIP'이면 총액의 15%, 'GOLD'이면 10%를 할인합니다.
2. 할인 적용 후 최종 금액이 50,000원 미만이면 배송비 3,000원을 추가합니다.
3. 50,000원 이상이면 배송비는 무료(0원)입니다.

```python
my_cart = [{
    "price": 25000, 
    "quantity": 1
    }, {"price": 15000, "quantity": 2}]

def calculate_checkout_price(my_cart, membership_grade):
    # 주어진 데이터
    shipping_fee = 3000

    # 할인률 계산
    discount_rate = 0
    if membership_grade == "GOLD":
        discount_rate = 0.1
    elif membership_grade == 'VIP':
        discount_rate = 0.15

    # 합계 계산 로직
    total = 0
    for item in my_cart:
        price = item['price']
        quantity = item['quantity']
        total += price * quantity

    # 할인 로직
    total = total * (1 - discount_rate)

    if total >= 50000:
        shipping_fee = 0

    total += shipping_fee
    return int(total)
```

```python
my_cart = [{
    "price": 25000, 
    "quantity": 1
    }, {"price": 15000, "quantity": 2}]

def calculate_discount_rate(membership_grade):
    
    discount_rate = 0
    if membership_grade == "GOLD":
        discount_rate = 0.1
    elif membership_grade == 'VIP':
        discount_rate = 0.15
    return discount_rate

def calculate_discount_rate(membership_grade):
    
    discount_rate_dict = {
        "GOLD" : 0.1,
        "VIP" : 0.15
    }
    discount_rate = discount_rate_dict.get(membership_grade, 0) # dic.get(key, default_value)

    return discount_rate

def calculate_checkout_price(my_cart, membership_grade):
    # 주어진 데이터
    shipping_fee = 3000
    discount_rate = calculate_discount_rate(membership_grade)

    # 합계 계산 로직
    total = 0
    for item in my_cart:
        price = item['price']
        quantity = item['quantity']
        total += price * quantity

    # 할인 로직
    total = total * (1 - discount_rate)

    if total >= 50000:
        shipping_fee = 0

    total += shipping_fee
    return int(total)
```

```python


my_cart = [{"price": 25000, "quantity": 1}, {"price": 15000, "quantity": 2}]
print(calculate_checkout_price(my_cart, "GOLD")) # 결과: 52500
```

### 문제 7
서버에서 받아온 다양한 형태의 데이터가 섞인 리스트에서 정수 ID만 추출하여 정형화하는 함수 `normalize_user_data`를 작성하세요.

**조건:**
1. 숫자형 문자열("105"), 숫자(101), 리스트 안의 숫자([102]) 등을 모두 정수로 변환하세요.
2. 중복된 ID는 제거하고, 결과는 오름차순으로 정렬된 리스트여야 합니다.

```python
dirty_data = ["105", 101, [102, 103], "101", 104, [105]]

# dirty_data = ["105", 101, "101", 104, ]

def normalize_user_data(dirty_data):
    clean_data = []

    for data in dirty_data:
        # data : str
        # data : int
        
        # data : list

        # 우리가 아는 것 : list인지 아닌지 구분해야 겠다.
        # 모르는 것 : list인지 아닌지 어떻게 구분하지?

        if type(data) == list:
            for item in data:
                clean_data.append(int(item))
        else:
            clean_data.append(int(data))


    # 중복 제거
    clean_data = set(clean_data)
    clean_data = list(clean_data)
    clean_data = sorted(clean_data)

    return clean_data
```

```python


dirty_data = ["105", 101, [102, 103], "101", 104, [105]]
print(normalize_user_data(dirty_data)) # 결과: [101, 102, 103, 104, 105]
```

### 문제 8
특정 조건을 만족하는 유저들을 추출하여 알림 대상자 리스트를 반환하는 함수 `get_notification_targets`를 작성하세요.

**조건:**
1. `is_active`가 `True`인 유저 중에서 선별합니다.
2. `last_login`이 7일 이상이거나, `point`가 10,000점 이상인 유저가 대상입니다.
3. 대상자의 'name'만 리스트에 담아 반환하세요.

```python
users_data = {
    "u1": {"name": "Alice", "is_active": True, "last_login": 10, "point": 5000},
    "u2": {"name": "Bob", "is_active": False, "last_login": 1, "point": 15000},
    "u3": {"name": "Charlie", "is_active": True, "last_login": 2, "point": 12000}
}
result = []
# 우리에게 필요한건 {name, isactice}등이 들어있는 dic
for user_data in users_data.values():
    if user_data['is_active']:
        if user_data['last_login'] >= 7 or user_data['point'] >= 10000:
            result.append(user_data['name'])
print(result)


def get_notification_targets(users_data):
    result = []
    # 우리에게 필요한건 {name, isactice}등이 들어있는 dic
    for user_data in users_data.values():

        is_active = user_data['is_active']
        last_login = user_data['last_login']
        point = user_data['point']
        name = user_data['name']


        if is_active:
            if last_login >= 7 or point >= 10000:
                result.append(name)
    return result


```

```python


users_data = {
    "u1": {"name": "Alice", "is_active": True, "last_login": 10, "point": 5000},
    "u2": {"name": "Bob", "is_active": False, "last_login": 1, "point": 1500},
    "u3": {"name": "Charlie", "is_active": True, "last_login": 2, "point": 12000}
}
print(get_notification_targets(users_data)) # 결과: ['Alice', 'Charlie']
```
