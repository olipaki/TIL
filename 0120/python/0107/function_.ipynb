```python
print('hello world')
# 기능 : 우리(개발자)에게 보여준다.
# 입력값 : () 사이에 들어가는 것들
# 반환값 : 없습니다.

total = sum([1, 2, 3, 4])
print(total)

max_value = max([1, 2, 3, 4])
print(max_value)

lst = [1, 2, 3, 4]
def my_max(lst):
    max_value = lst[0]

    for i in lst:
        if i > max_value:
            max_value = i
    return max_value
print(my_max([1, 2, 3, 4]))
```

```python
# 어떤 숫자를 받아서 제곱하는 함수를 만들고 싶어요.

x = 10
result = x ** 2

def square(n):
    result = n ** 2
    return result

value = square(12)
value = square(11)

print(value)
```

```python
# x, y 사이의 정수의 합
x = 10
y = 3
# x를 y보다 작게 만들고 싶다.
# if x > y:
#     x, y = y, x


def order_two_number(num1, num2):
    # 앞에 나오는 숫자가 작게 만들고 싶어.
    if num1 > num2:
        num1, num2 = num2, num1

    return num1, num2
# 함수명이 충분해서, 안의 내용을 볼 필요가 없어요.
small, big = order_two_number(x, y) # 필요성을 느꼈어.

print(sum(range(small, big)))
```

```python
value = print(1)
print(value) # 결과물이 필요 없는 함수도 있다.
# 결과물이 없음을 None으로 표현합니다.

# 입력값이 없을수도 있긴 합니다.
def func():
    return 'hello world'

value = func()
print(value)

# return 자체가 없을수도 있어요

def say_hi():
    print('hi')

say_hi() # return이 생략되면 'None'을 return하는 것이랑 같습니다.
```

```python
# return이 2개 사용 가능합니다.
# 다만, return은 1번만 됩니다.
# 추가로, return은 함수를 끝내는 역할을 합니다.
def is_one(n):
    if n == 1:
        return True
    else:
        return False
    
```

```python
# print(1, 2, 3, 4)
# print는 여러개의 인자를 받습니다.

def my_sum(*args):
    total = 0

    for num in args:
        total += num
    return total

my_sum(1, 2, 3, 4, 5, 1, 2, 3, 4, )
```

```python
def show_menu(**kwargs):
    print(kwargs)

show_menu(a=1, b=2, c=3)
```

```python
# 기본값을 넣을 수 있음.
def add_number(a, b, c=100):
    return a + 2*b + 3*c

print(add_number(1, 2, 3))
print(add_number(1, 2))

# 함수를 호출할 때 매개변수를 명시할 수 있어요ㅣ
print(add_number(b=1, a=2, c=3))
```

```python
# problem6 문제5 발췌
scores = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}
students = {}

def get_status_by_score(score):
    if score >= 80:
        return 'pass'
    else:
        return 'fail'

for name, score in scores.items():
    # students라는 새로운 dictionary에 score에 따른 상태를 넣고싶어.

    students[name] = get_status_by_score(score)
```
