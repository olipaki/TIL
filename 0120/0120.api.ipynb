```python
# 1. `시도별 실시간 측정정보 조회`에서  확인가능한 시도 이름을 전부 작성하시오.
#     1. `서울`인지 `서울특별시`인지 등 parameter를 나열하세요



import requests

# 1. 설정 정보
URL = "http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty"
SERVICE_KEY = "b9a0b7aeeffea832f768cd790733382b77e828553f10afdad9924188a5c8a741" # Decoding 키를 추천합니다.

params = {
    'serviceKey': SERVICE_KEY,
    'returnType': 'json',
    'numOfRows': '100',
    'pageNo': '1',
    'sidoName': '서울', # 위 리스트 중 선택 (예: 제주, 경기 등)
    'ver': '1.0'
}



try:
    # 2. API 호출
    response = requests.get(URL, params=params)
    data = response.json()
    items = data['response']['body']['items']

    # 3. 자료구조 재구성 (stationName으로 즉시 접근 가능한 Dictionary)
    # [문제 해결] 서울 데이터를 측정소 이름으로 접근하기 쉽게 변환
    seoul_data_dict = {item['stationName']: item for item in items}

    # 4. 초미세먼지(pm25Value) 농도가 가장 낮은 곳 찾기
    # 숫자가 아닌 데이터('-')가 섞여 있을 수 있어 필터링이 필요합니다.
    valid_items = [item for item in items if item['pm25Value'] and item['pm25Value'].isdigit()]
    
    if valid_items:
        # pm25Value를 기준으로 오름차순 정렬
        min_station = min(valid_items, key=lambda x: int(x['pm25Value']))
        
        print(f"--- 조회 결과 ---")
        print(f"초미세먼지 농도가 가장 낮은 측정소: {min_station['stationName']}")
        print(f"농도: {min_station['pm25Value']}㎍/㎥")
        print(f"------------------")
        
        # 5. 자료구조 사용 예시 (종로구 데이터만 쏙 빼오기)
        if '종로구' in seoul_data_dict:
            jongno = seoul_data_dict['종로구']
            print(f"종로구 현재 미세먼지(PM10): {jongno['pm10Value']}")

except Exception as e:
    print(f"에러가 발생했습니다: {e}")


```
