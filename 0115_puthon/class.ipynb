```python
# ## 자판기

# - 사람이 자판기에서 물건을 구입합니다
# - 현금결제 카드결제가 가능합니다
# menu
# snack /water /glossary

# Cash - change
# credit_card - accept

vending = j_machine()
class j_machine : 
    def __init__(self):
        # 1. 자판기가 팔 물건과 가격 (메뉴판)
        self.menu = {"스낵" : 2000,"물" : 1000, "생필품" : 3000}
# 2. 자동 준비: 자판기를 만들자마자(machine = J_machine()) 자동으로 0원 세팅 완료!
        self.input_money = 0
        print("오픈")

    def cash_on(self,money):
     # 현금을 넣으면 '금액(self.input_money)'추가
        self.input_money += money
        print(f"현재 {self.input_money}원 넣었습니다.")
        # print("구매가능")

        for name,price in self.menu.items():
            if self.input_money >= price :
                print(f"현재{name},{price} 구입할 수 있습니다.")
# **첫 번째 자리(name)**에는 딕셔너리의 **키(Key, 물건 이름)**를 넣어주고,
# **두 번째 자리(price)**에는 딕셔너리의 **값(Value, 가격)**을 넣어주기로 약속되어 있어요.
    
    #거스름돈 필요 (현금일 경우 10000원만 사용가능 / 거스름돈 1000원)
    def buy_item(self,item_name):
        price = self.menu[item_name]
        if self.input_money >= price : # 손님넣은 돈
            self.input_money -= price 
            print(f"{item_name}드립니다.")
            print(f"남은잔액은 {self.input_money}입니다.")
        else :
            print(f"{price - self.input_money}원 부족합니다.")


vending.cash_on(5000)
        



```
