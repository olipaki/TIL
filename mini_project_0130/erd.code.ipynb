````python
| name | 강좌명 |
| --- | --- |
| major_name | 전공 |
| teacher | 강사 |
| teacher/name | 강사 / 이름 |
| teacher/gender | 강사 / 성별 |
| start_day | 강의 시작일자 |
| end_day | 강의 종료일자 |

### 예시 응답

```json
[
		{
				"name" : "한국사완전정복",
				"major_name" : "AI학부",
				"teacher" : {
						"user_no" : 5,
						"name" : "alex",
						"gender" : "여자"
				},
				"start_day" : "2026-01-22",
				"end_day" : "2026-05-12"
		}	,
		{
				"name" : "한국사기초",
				"major_name" : "AI학부",
				"teacher" : {
						"id" : 6,
						"name" : "김진우",
						"gender" : "남자"
				},
				"start_day" : "2026-02-22",
				"end_day" : "2026-07-12"
		}	,
]
```

## SQL

```sql
select co.name, ma.major_name, us.user_no, us.name, us.gender, co.start_day, co.end_day
from courses co
join major ma
on co.major_no = ma.major_no
join teacher te
on co.teacher_no = te.teacher_no
join user us
on te.user_no = us.user_no
where ma.major_no = 5
```
````
