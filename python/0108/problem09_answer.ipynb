### 문제 1
여러 명의 학생 성적 데이터가 담긴 복합 리스트입니다. 각 학생의 `grades` 리스트에서 80점 이상의 점수만 추출하여 평균을 계산하고, 그 평균이 90점 이상인 학생의 이름만 `top_students` 리스트에 담으세요.

```python
students = [
    {"name": "John", "grades": [85, 92, 78, 100]},
    {"name": "Sarah", "grades": [88, 95, 90, 89]},
    {"name": "Tyler", "grades": [70, 75, 82, 80]},
    {"name": "Emma", "grades": [95, 98, 88, 92]},
    {"name": "Emma2", "grades": [1, 2, 3, 4]},
]
# 각각의 학생들에 대해서
# 점수가 80점 이상인 것들만 확인해서 평균을 구하겠다.
# 90이 넘으면 리스트에 담겠다.

# 점수가 80점 이상인 것들만 확인해서 평균을 구하는 함수.
def get_average(grades):
    count = 0
    total = 0
    for grade in grades:
        if grade >= 80:
            count +=1
            total += grade

    if count == 0:
        return None            
    
    return total / count
            
def get_average(grades):
    grades = [grade for grade in grades if grade >= 80]

    if len(grades) == 0:
        return None
    return sum(grades) / len(grades)


top_students = []
for student in students:
    name = student['name']
    grades = student['grades']
    average_over_80 = get_average(grades)

    if average_over_80 == None:
        continue

    if average_over_80 >= 90:
        top_students.append(name)
print(top_students)
```

### 문제 2
소셜 미디어 API 응답 데이터 형태입니다. 각 유저(`users`)가 작성한 게시글(`posts`)들 중, 좋아요(`likes`) 수가 50개 이상인 게시글의 제목(`title`)만 모두 모아 하나의 리스트로 만드세요.

```python
api_response = {
    "status": "success",
    "data": {
        "users": [
            {
                "username": "user1",
                "posts": [
                    {"title": "오늘의 식단", "likes": 30},
                    {"title": "여행 사진", "likes": 150}
                ]
            },
            {
                "username": "user2",
                "posts": [
                    {"title": "코딩 공부법", "likes": 88},
                    {"title": "운동 일지", "likes": 10}
                ]
            },
            {
                "username": "user3",
                "posts": [
                    {"title": "ai 공부법", "likes": 808},
                    {"title": "일기", "likes": 100}
                ]
            },
        ]
    }
}

users = api_response['data']['users']

popular_titles = []
for user in users:
    # user : dictionary
    posts = user['posts']

    # posts: list
    for post in posts:
        title = post['title']
        likes = post['likes']

        if likes >= 50:
            popular_titles.append(title)
        

# 코드를 작성하세요

print(f"인기 게시글 제목: {popular_titles}")
```

```python
from collections import defaultdict

products = [
    {"category": "전자제품", "name": "키보드"},
    {"category": "의류", "name": "티셔츠"},
    {"category": "전자제품", "name": "마우스"},
    {"category": "전자제품", "name": "노트북"},
    {"category": "식품", "name": "사과"},
    {"category": "식품", "name": "배"},
    {"category": "의류", "name": "청바지"}
]


grouped_data = defaultdict(list)

# 이후를 위한 데이터 전처리
for product in products:
    category = product['category']
    name = product['name']

    grouped_data[category].append(name)
    

# 코드를 작성하세요
grouped_data = dict(grouped_data)
print(f"그룹화 결과: {grouped_data}")
```
