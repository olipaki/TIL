# 슬라이싱

list를 잘라서 새로운 list를 만드는 방법

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
print(lst[2:6])
# 인덱스가 생략되면 '전체'의 의미입니다.
# 앞의 생략되면 -> 처음부터 / 뒤가 생략되면 -> 끝까지.
print(lst[:6])
print(lst[3:])
# 얕은 복사.
print(lst[:])

# step
print(lst[::2])
print(lst[::-1])


```

# 멤버십 연산자

list에 값이 들어있는지를 확인하는 연산자

```python
print(1 in [1, 2, 3, 4])
print(10 in [1, 2, 3, 4])

print(10 not in [1, 2, 3, 4])
```

```python
lst = ['a', 'b', 'c', 'd']

for i in range(len(lst)):
    el = lst[i]
    # 인덱스 : i / 값 : el

for i, el in enumerate(lst):
    print(i, el)
```

# 리스트의 메서드들(함수들)

```python
# 값을 삽입하는 메서드
lst = [1, 2, 3, 4]

lst.append(5)

print(lst)
```

```python
# 값을 삭제하는 메서드
lst = [1, 2, 3, 4]

popped_el = lst.pop()

print(lst)
print(popped_el)

lst = [1, 2, 3, 4]

lst.pop(2) # index를 지정해서 삭제
print(lst)
```

# 리스트 컴프리헨션

리스트를 간편하게 만드는 방법 (규칙에 따라서)

```python
# lst = [1, 2, 3, 4, 5, 6]
lst = []
for num in range(1, 7):
    lst.append(num)

print(lst)
```

```python
lst = [num for num in range(1, 7)]
print(lst)
```

```python
# 짝수로만 만들어진 list
lst = []
for num in range(1, 21):
    if num % 2 == 0:
        lst.append(num)

print(lst)
```

```python
lst = [num for num in range(1, 21) if num % 2 == 0]
print(lst)
```
