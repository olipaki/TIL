# while


```python
# while condition:
#     반복할 코드

# if condition:
#     코드
n = 10
while n > 0 :
    print(n)
    n -= 1

```

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
lst = ['a', 'b', 'c']
# while 문을 활용해서 a b c d를 출력해봅시다.
# lst[0]
# lst[1]
# lst[2]
# lst[3]
i = 0
# while i <= len(lst)-1:
while i < len(lst):
    print(lst[i])
    i += 1

    print(i)
```

```python
# while문은 어떻냐면

# - 조건이 True일때 실행시켜.
# - 그냥 실행시켜. 근데 조건이 True면 멈춰

n = 0
while n < 10:
    print(n)
    n += 1
print()

n = 0
while True:
    print(n)
    n += 1
    if n >= 10:
        break
```
