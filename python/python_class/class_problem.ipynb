# 파이썬 클래스 연습문제

### 문제 1
이름(`name`)과 학번(`student_id`)을 속성으로 가지는 `Student` 클래스를 작성하세요.
1. `__init__` 생성자를 사용하여 두 속성을 초기화하세요.
2. 본인의 이름과 학번을 가진 인스턴스를 생성하여 각 속성값을 출력하세요.

```python
# 여기에 코드를 작성하세요.
class Student:
    # 클래스 변수의 공간
    school_name = "새싹 도봉캠"

    def __init__(self, student_name, student_id):
        # 인스턴스 변수의 공간1
        self.name = student_name
        self.student_id = student_id

    def say_hi(self):
        print(f"안녕 나는 {self.name}. {self.school_name} 출신이야")
student1 = Student("강성신", 10)
student2 = Student("김아네스", 9)
print(student1.name)
print(student1.student_id)

print(student1.school_name)
print(student2.school_name)
student1.say_hi()
```

### 문제 2
계산기 기능을 수행하는 `Calculator` 클래스를 만드세요.
1. 생성자에서 초기값 0을 가지는 `result` 인스턴스 변수를 생성합니다.
2. `add(self, num)` 메서드를 정의하여 호출 시 `result`에 값을 더합니다.
3. `subtract(self, num)` 메서드를 정의하여 호출 시 `result`에서 값을 뺍니다.
4. 메서드 호출 후 현재의 `result`를 확인하는 코드를 작성하세요.

```python
# 여기에 코드를 작성하세요.

class Calculator:

    def __init__(self, num = 0):
        self.result = num
        

    def add(self, num):
        self.result += num
        return self.result
    
    def subtract(self, num):
        self.result -= num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()

cal1.add(10)
cal1.add(20)
cal1.subtract(5)
print(cal1.result)
print(cal2.result)
```

### 문제 3
매직 메서드를 사용하여 도서 정보를 관리하는 `Book` 클래스를 만드세요.
1. 생성자에서 제목(`title`)과 페이지 수(`pages`)를 입력받아 저장합니다.
2. `__str__` 메서드를 구현하여 `print()` 함수 호출 시 "제목: [제목]"이 출력되게 하세요.
3. `__len__` 메서드를 구현하여 `len()` 함수 호출 시 페이지 수가 반환되게 하세요.

```python
# 여기에 코드를 작성하세요.
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return self.title
    def __len__(self):
        return self.pages
book = Book('해리포터', 100)
print(book)
print(len(book))
```
